## Ho$_2$Ti$_2$O$_7$

```python
HTO=CEF.CEFmodel('Ho3+') 

HTO.PC_variable = 'R1', 'R2', 'Theta', 'q1', 'q2'
HTO.clearPC()
HTO.addPC('R1, 0, 0, q1')
HTO.addPC('R1, 180, 0, q1')
HTO.addPC('R2, 180-Theta,   0, q2')
HTO.addPC('R2,     Theta,  60, q2')
HTO.addPC('R2, 180-Theta, 120, q2')
HTO.addPC('R2,     Theta, 180, q2')
HTO.addPC('R2, 180-Theta, 240, q2')
HTO.addPC('R2,     Theta, 300, q2')

HTO.PC_value_ini=np.array([1.594, 1.594, 79.4, 0.5, 0.333])
HTO.FWHM=0

HTO.Evaluate_pattern(Ei=80, Temperature=5, Plotcontrol=True, UsePCini=True);
print('Eigen-energies =',HTO.eigenval)
print('Akq (meV) =\n',CEF.BkqtoAkq(HTO.Bkq,'Ho3+'))
HTO.gtensor()
```

```python
# Read Wybourne Bkq from reference
HTO=CEF.CEFmodel('Ho3+')
HTO.readBkq("Ho2Ti2O7_Ref.dat", convention='Wybourne')
print('Akq(meV)\n',CEF.BkqtoAkq(HTO.Bkq,'Ho3+'))

HTO.Evaluate_pattern(Ei=80, Temperature=5, Plotcontrol=True);
HTO.gtensor()
```

## Er$_2$Ti$_2$O$_7$


```python
ETO=CEF.CEFmodel('Er3+') 

ETO.PC_variable = 'R1', 'R2', 'Theta', 'q1', 'q2'
ETO.clearPC()
ETO.addPC('R1, 0, 0, q1')
ETO.addPC('R1, 180, 0, q1')
ETO.addPC('R2, 180-Theta,   0, q2')
ETO.addPC('R2,     Theta,  60, q2')
ETO.addPC('R2, 180-Theta, 120, q2')
ETO.addPC('R2,     Theta, 180, q2')
ETO.addPC('R2, 180-Theta, 240, q2')
ETO.addPC('R2,     Theta, 300, q2')

ETO.PC_value_ini=np.array([1.571, 1.571, 79.1, 0.5, 0.333])
ETO.FWHM=0

ETO.Evaluate_pattern(Ei=100, Temperature=5, Plotcontrol=True, UsePCini=True);
print('Eigen-energies =',ETO.eigenval)
print('Akq (meV) =\n',CEF.BkqtoAkq(ETO.Bkq,'Er3+'))
ETO.gtensor()
```

```python
Comounds= 'Er2Ge2O7', 'Er2Ti2O7', 'Er2Pt2O7', 'Er2Sn2O7'
Akq=np.array([[39.3, 36.2, 275, 1.23, -19.10, 26.6],  #Er2Ge2O7 
              [37.5, 33.5, 282, 1.25, -17.15, 21.6],  #Er2Ti2O7 
              [49.0, 31.2, 262, 1.10, -19.15, 24.6],  #Er2Pt2O7 
              [50.8, 28.4, 266, 1.10, -17.15, 20.9]])  #Er2Sn2O7 
r2,r4,r6 = 0.7111, 1.309, 4.816   # <r^2>,  <r^4>, <r^6
Akq[:,0]=Akq[:,0]*r2
Akq[:,1:3]=Akq[:,1:3]*r4
Akq[:,3:6]=Akq[:,3:6]*r6
print('             A_2^0   A_4^0   A_4^3   A_6^0   A_6^3   A_6^6')
for i, term in enumerate(Akq):
    print(Comounds[i],term)

ETO.readBkq("Er2Ti2O7_Ref.dat", printcontol=False)  # read real Akq from file
ETO.Bkq = CEF.AkqtoBkq(ETO.Bkq, 'Er3+') # convert Akq to Bkq
ETO.Evaluate_pattern(Ei=100, Temperature=5, Plotcontrol=True);
print('\n Eigen-energies for Er2Ti2I7:\n',ETO.eigensys()[0])
```
