```python
import sys
import copy
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
sys.path.append("/Users/dunzhiling/Documents/Research/Program/Python/PointChargeFit/arXivProgram/")
import PointChargeCEF_Oct2020 as CEF

```

## Tm-tripod

```python
TMSO=CEF.CEFmodel('Tm3+')

TMSO.PC_variable = 'R1', 'R2', 'R3', 'Theta2', 'Theta3', 'Phi', 'q1', 'q2', 'q3'

TMSO.clearPC()
TMSO.addPC('R1, 0, 0, q1')
TMSO.addPC('R1, 180, 0, q1')
TMSO.addPC('R2, Theta2, Phi, q2')
TMSO.addPC('R2, Theta2, 360-Phi, q2')
TMSO.addPC('R2, 180-Theta2, 180-Phi, q2')
TMSO.addPC('R2, 180-Theta2, 180+Phi, q2')
TMSO.addPC('R3, 180-Theta3, 0, q3')
TMSO.addPC('R3, Theta3, 180, q3')


TMSO.PC_value_ini = np.array([1.64, 1.57, 1.48, 78, 76.5, 59, 0.5 ,0.3, 0.15]) 
TMSO.Evaluate_pattern(Ei=50, Plotcontrol=True, UsePCini=True)

print(TMSO.eigenval)
TMSO.diagonalG()
```

## Dy-tripod

```python
DMSO=CEF.CEFmodel('Dy3+')

DMSO.PC_variable = 'R1', 'R2', 'R3', 'Theta2', 'Theta3', 'Phi', 'q1', 'q2', 'q3'

DMSO.clearPC()
DMSO.addPC('R1, 0, 0, q1')
DMSO.addPC('R1, 180, 0, q1')
DMSO.addPC('R2, Theta2, Phi, q2')
DMSO.addPC('R2, Theta2, 360-Phi, q2')
DMSO.addPC('R2, 180-Theta2, 180-Phi, q2')
DMSO.addPC('R2, 180-Theta2, 180+Phi, q2')
DMSO.addPC('R3, 180-Theta3, 0, q3')
DMSO.addPC('R3, Theta3, 180, q3')


DMSO.PC_value_ini = np.array([1.68, 1.57, 1.38, 78, 76.5, 59, 0.5 ,0.3, 0.15]) 

DMSO.Evaluate_pattern(Ei=80, Plotcontrol=True, UsePCini=True)

print(DMSO.eigenval)
DMSO.diagonalG();
```
