The site-disordered R3+ ion has the same point charge environment like RMgGaO$_{4}$, but surrounded by O2
R-O distance = 2.1 A 

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
sys.path.append("/Users/dunzhiling/Documents/Research/Program/Python/PointChargeFit/arXivProgram/")
import PointChargeCEF_Oct2020 as CEF

HMGO = CEF.CEFmodel('Tm3+')   # creast a CEFmodel object named TMGO

def octhedron(Element):
    SiteDisorder=CEF.CEFmodel(Element)
    SiteDisorder.PC_variable = 'R','Theta','q'
    # then add 6 Point charge in spherical coordination
    SiteDisorder.addPC('R, Theta, 60, q')
    SiteDisorder.addPC('R, Theta, 180, q')
    SiteDisorder.addPC('R, Theta, 300, q')
    SiteDisorder.addPC('R, 180-Theta, 0, q')
    SiteDisorder.addPC('R, 180-Theta, 120, q')
    SiteDisorder.addPC('R, 180-Theta, 240, q')
    return SiteDisorder
```

## Ho$^{3+}$

```python
Ho_SD=octhedron('Ho3+')
Ho_SD.PC_value_ini=np.array([2.1*0.72, 57, 0.3])
Ho_SD.Evaluate_pattern(Ei=80, Temperature=5, Plotcontrol=True, UsePCini=True);
print(Ho_SD.eigenval)
```
